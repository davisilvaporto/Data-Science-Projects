```python
#função para gerar gráfico de correlação dos dados 
def plot_corr(df, size=7):
  corr= df.corr()
  fig, ax = plt.subplots(figsize=(size, size))
  ax.matshow(corr)
  plt.xticks(range(len(corr.columns)), corr.columns)
  plt.yticks(range(len(corr.columns)), corr.columns)
```

```python
#plotando gráficos de correlação
plot_corr(df)
```

```python
#plotando tabela de correlação
df.corr()
```
