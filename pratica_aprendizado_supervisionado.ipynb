## importar o pacote de modelos supervisionado baseados em vizinhança do sklearn

```
from sklearn import neighbors
```

## Importar uma base de dados disponível no sklearn

```
from sklearn.datasets import load_digits
```

## Carregar os inputs e outputs da base

```
# https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_digits.html
X, y = load_digits(return_X_y=True)
```

## Instanciando e treinando um KNN

```
model = neighbors.KNeighborsClassifier()

X_train = X[:int(.9 * len(X))]
y_train = y[:int(.9 * len(y))]
X_test = X[int(.9 * len(X)):]
y_test = y[int(.9 * len(y)):]

model.fit(X_train, y_train)
```

```
model.score(X_train, y_train)
```

## Calculando o score do modelo treinado

```
model.score(X_test, y_test)
```
