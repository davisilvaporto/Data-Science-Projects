```
def substituirPadraoRegex(variavel,padraoRegex,substituto):
  if(type(variavel)==str):
    return padraoRegex.sub(substituto,variavel)
  else:
    return [padraoRegex.sub(substituto,texto) for texto in variavel]
```

## Estimando a perplexidade de textos em relação aos modelos de linguagem

```
def treinarModeloLinguagem(lista_textos):
  todas_palavras = ' '.join(lista_textos)
  tokens_todas_palavras = WhitespaceTokenizer().tokenize(todas_palavras)
  everygrams, vocabulario = padded_everygram_pipeline(2,tokens_todas_palavras)
  modelo = MLE(2)
  modelo.fit(everygrams, vocabulario)
  return modelo
```

```
def calcularPerplexidade(modelo, texto):
  perplexidade = 0
  tokens = WhitespaceTokenizer().tokenize(texto)
  tokens_fakechar = [list(pad_both_ends(token,n=2)) for token in tokens]
  bigramas_tokens = [list(ngrams(token,2)) for token in tokens_fakechar]

  for bigramas_token in bigramas_tokens:
    perplexidade += modelo.perplexity(bigramas_token)

  return perplexidade
```

# Tema 5 - Aprimoramento de modelo de linguagem usando Laplace.

## Modelo de Laplace para lidar com perplexidade infinita

```
from nltk.lm import Laplace

def treinarModeloLinguagemLaplace(lista_textos):
  todas_palavras = ' '.join(lista_textos)
  tokens_todas_palavras = WhitespaceTokenizer().tokenize(todas_palavras)
  everygrams, vocabulario = padded_everygram_pipeline(2,tokens_todas_palavras)
  modelo = Laplace(2)
  modelo.fit(everygrams, vocabulario)
  return modelo
```

```
modelo_ing_Laplace = treinarModeloLinguagemLaplace(ingles_treino)
```

```
print(calcularPerplexidade(modelo_ing_Laplace,port_teste.iloc[0]))
```

## Classificando textos com modelos de linguagem

```
def avaliaIdioma(lista_textos):
  idiomas=[]
  for texto in lista_textos:
    perp_portugues = calcularPerplexidade(modelo_port_Laplace,texto)
    perp_ingles = calcularPerplexidade(modelo_ing_Laplace,texto)
    perp_espanhol = calcularPerplexidade(modelo_esp_Laplace,texto)
    if perp_ingles >= perp_portugues <= perp_espanhol:
      idiomas.append("portugues")
    elif perp_portugues > perp_ingles < perp_espanhol:
      idiomas.append("ingles")
    else:
      idiomas.append("espanhol")
  return idiomas
```
